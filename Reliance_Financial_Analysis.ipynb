```python
import pandas as pd

data = pd.read_csv("Reliance_1.csv", skiprows=1)
data

```

```python
cols = ["Revenue","Net Profit","Current Assets","Current Liabilities","Total Debt","Equity"]

for c in cols:
    data[c] = data[c].astype(str).replace(",","", regex=True).astype(float)

data

```

```python
data["Profit_Margin"] = data["Net Profit"] / data["Revenue"]
data["Current_Ratio"] = data["Current Assets"] / data["Current Liabilities"]
data["Debt_Equity"] = data["Total Debt"] / data["Equity"]

data

```

```python
data

```

```python
data["Profit_Margin"] = (data["Profit_Margin"] * 100).round(2)
data
```

```python
import matplotlib.pyplot as plt

plt.plot(data["Year"], data["Profit_Margin"])
plt.title("Profit Margin Trend")
plt.xlabel("Year")
plt.ylabel("Profit Margin (%)")
plt.show()

```

```python
plt.plot(data["Year"], data["Current_Ratio"])
plt.title("Current Ratio Trend")
plt.xlabel("Year")
plt.ylabel("Current Ratio")
plt.show()

```

```python
plt.plot(data["Year"], data["Debt_Equity"])
plt.title("Debt-Equity Ratio Trend")
plt.xlabel("Year")
plt.ylabel("Debt-Equity")
plt.show()

```
